```python
#Creación de spider o crawler que entra a paginas web y extrae información
from bs4 import BeautifulSoup
import requests, re, time

pagina = requests.get("https://es.wikipedia.org/wiki/Python")
soup = BeautifulSoup(pagina.content, 'html.parser')

listalinksexplorados = set()
listalinksporexplorar = set()

baselink = "https://es.wikipedia.org"
```

```python
#Extraemos los links de la pagina
for link in soup.findAll("a"):
    if str(link.get("href"))[0] == "/" and str(link.get("href"))[1] != "/" and baselink + str(link.get("href")) not in listalinksexplorados:
        listalinksporexplorar.add(baselink + str(link.get("href")))
```

```python
while(len(listalinksporexplorar) != 0):
    link = listalinksporexplorar.pop()
    pagina = requests.get(link)
    soup = BeautifulSoup(pagina.content, 'html.parser')
    #Si quisieramo extraer algo de las paginas visitadas, aquí usaríamos soup.find(..)
    print("Links Explorados: " + str(len(listalinksexplorados)) + " Links a Explorar: " + str(len(listalinksporexplorar)))
    listalinksexplorados.add(link)
    time.sleep(1) #Para que no se vea tan rapido el proceso y no nos bloquee la pagina
    for link in soup.findAll("a"): #Extraemos los links de c/u de las paginas, date cuenta que vamos agregando todos los links de las pags conectadas a estas últimas
        if str(link.get("href"))[0] == "/" and str(link.get("href"))[1] != "/" and baselink + str(link.get("href")) not in listalinksexplorados:
            listalinksporexplorar.add(baselink + str(link.get("href")))
```
