```python
from sklearn.model_selection import train_test_split, KFold
from sklearn.metrics import r2_score
from sklearn import linear_model
from bs4 import BeautifulSoup
import requests, re, time
import pandas as pd
import numpy as np

pagina = requests.get('https://es.wikipedia.org/wiki/Anexo:Países_por_PIB_histórico_(nominal)')
soup = BeautifulSoup(pagina.content, 'html.parser')
```

```python
tabla=soup.find_all('table')[1]
```

```python
tabla
```

```python
aiura=[]
for i in range(10):
    aiura.append(str(i))

nombrecolumnas = []
for th in tabla.find_all('th'):
    if str(th.string)[0]=="2":
        if th.string[-1] in aiura:
            nombrecolumnas.append(th.string)
            print(th.string)
        else:
            nombrecolumnas.append(th.string[:-1])
            print(th.string[:-1])
        

nombrecolumnas
```

```python
for elem in tabla.find_all('td'):
    print(elem.string)
```

```python
paises=[]
for elem in tabla.find_all('a'):
    if elem.string!=None:
        paises.append(elem.string)
paises
```

```python
valores =[]
i=1
for elem in tabla.find_all('td'):
    if elem.string!=None and elem.string != "" and elem.string != str("\n") and elem.string != str(i)+"\n" and elem.string != str(i):
        if elem.string[-1] == "n" and elem.string.__len__()>3:
            aiura = elem.string[:-1].replace(",","")
            aiura= int(aiura.replace(".",""))
            valores.append(aiura)
            i+=1
            if(i>=20*13):
                break
        elif elem.string.__len__()>3:
            aiura = elem.string.replace(",","")
            aiura= int(aiura.replace(".",""))
            valores.append(aiura)
            i+=1
            if(i>20*13):
                break
```

```python
len(valores)
```

```python
paises2 = []
for i in range(20):
    paises2.append(paises[i])
paises2
```

```python
valores = np.array(valores)
valores.shape=(len(paises2),len(nombrecolumnas))
valores
```

```python
df = pd.DataFrame(valores, columns=nombrecolumnas) # index=paises2 para poner los nombres de los paises
df
```

```python
X=df.drop(["2018", "2017","2010"], axis=1)
y=df["2010"]
```

```python
kf = KFold(n_splits=5)
kf.get_n_splits(X)

regr = linear_model.LinearRegression()
resultados = []
for train_index, test_index in kf.split(X):
    X_train, X_test = X.loc[train_index], X.loc[test_index]
    y_train, y_test = y.loc[train_index], y.loc[test_index]
    regr.fit(X_train, y_train)
    y_pred = regr.predict(X_test)
    print("R2: ", r2_score(y_test, y_pred))
    resultados.append(r2_score(y_test, y_pred))
print("R2: mean", np.mean(resultados))
```
